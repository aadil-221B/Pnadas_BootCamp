```python
import numpy as np
import pandas as pd
```

```python
# some python lists
names = ['Olga', 'Andrew', 'Brian', 'Telulah', 'Nicole', 'Tilda']
ages = [29, 21, 45, 23, 39, 46]
married = [False, True, True, True, False, True]
```

```python
pd.DataFrame({"name": names,
              "ages": ages,
              "married": married
             
             })
```

```python
{k:v for k,v in enumerate(names)}
```

```python
def convert_list_to_dict(l):
    return {k:v for k,v in enumerate(l)}
```

```python
dict_names = convert_list_to_dict(names)
dict_ages = convert_list_to_dict(ages)
dict_married = convert_list_to_dict(married)
```

```python
[[name, age, married]for name,age,married in zip(names, ages, married)]
```

```python
list_dict = [{"name": name, "age": age, "married": married}for name,age,married in zip(names, ages, married)]
```
