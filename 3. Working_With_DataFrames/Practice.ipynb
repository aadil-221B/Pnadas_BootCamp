## Merging Units With Columns Names

```python
import pandas as pd
import numpy as np
dataurl = 'https://andybek.com/pandas-nutrition'
```

```python
nutrition = pd.read_csv(dataurl,index_col="name")
nutrition.drop("Unnamed: 0", axis=1,inplace=True)
```

```python
nutrition.head()
```

```python
nutrition.sample(frac=0.15,axis=1).head()
```

```python
nutrition.replace(to_replace="[^a-zA-Z\s]", value="", regex=True)
```

```python
# we need to replace the numeric value too
units = nutrition.astype(str).replace(to_replace="[^a-zA-Z]", value="", regex=True)
units
```

```python
units = units.mode()
```

```python
units
```

```python
# cleaning units data
# find na vaalues
# find missing values
units.isna().any().any() # no na values
units = units.replace(to_replace="",value=pd.NA).dropna(axis=1)
```

```python
units
```

```python
for cols in units.columns:
    print(units[cols][0])
```

```python
 # make a dictionary of col's and respective units
```

```python
units.to_dict(orient="records")
```

```python
# another method
# dict comprehension

mapper = {cols: units[cols][0] for cols in units.columns}
```

```python
mapper
```

```python
# we want labels as columnnames_units
mapper_cols = {cols: cols+"_"+units[cols][0] for cols in units.columns}
mapper_cols
```

```python
# rename colums
nutrition.rename(columns=mapper_cols)
```
