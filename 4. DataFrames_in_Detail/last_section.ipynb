### assign()

```python
df_num = soccer.select_dtypes(include="number").iloc[:5,:5]
```

```python
df_num
```

```python
df_num.assign(
    sales_value = df_num["market_value"]/df_num["fpl_value"],
    sales_percent = (df_num["market_value"]/df_num["fpl_value"])/df_num["page_views"]
    
    
)
```

```python
df_num.assign(
    sales_value = lambda x: x["market_value"]/x["fpl_value"],
    sales_percent = lambda x: x["sales_value"]/x["page_views"]
    
    
)
```
